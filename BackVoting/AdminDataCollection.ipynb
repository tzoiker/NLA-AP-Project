```python
#Back Voting
import numpy as np
AllData = []
ins = open("AllData.txt", "r" )
for line in ins:
    AllData.append(line.replace('\n',''))

admin = []
adminNames = open("AdminNames.txt", "r" )
for line in adminNames:
    admin.append(line.replace('\n','').replace("(page does not exist)","").strip())
adminNames = open("FormerAdminNames.txt", "r" )
for line in adminNames:
    admin.append(line.replace('\n','').replace("(page does not exist)","").strip())

VotingList = []

for adName in admin:
    adName = adName[:adName.find(",")]
    adData = adName + "-->"
    for data in AllData:
        d = data.split(";")
        src = d[0]
        trg = d[1]
        year = d[4]
        if trg == adName:
            for aName in admin:
                aName = aName[:aName.find(",")]
                if aName == src:
                    adData = adData + src +":"+year + ";"
    VotingList.append(adData)

f = open("AdminAdminVotingData.txt","w")

for v in VotingList:
    f.write(v+"\n")
f.close()
```

```python
from __future__ import division
#VotingRatio
VotingData = []
Voting = open("AdminVotingData.txt", "r" )
for line in Voting:
    VotingData.append(line.replace('\n',''))
AAVotingData = []
AAVoting = open("AdminAdminVotingData.txt", "r" )
for line in AAVoting:
    AAVotingData.append(line.replace('\n',''))

VotingStats = []

for i in xrange(0,len(VotingData)):
    AV = VotingData[i]
    AAV = AAVotingData[i]
    if AV[AV.find(">")+1:] == "":
        AVStats = 0
    else:
        AVStats = len(AV[AV.find(">")+1:].split(";"))
    if AAV[AAV.find(">")+1:] == "":
        AAVStats = 0
    else:
        AAVStats = len(AAV[AAV.find("<")+1:].split(";"))
    percentage = 0
    if AVStats == 0:
        percentage = 0
    else:
        percentage = AAVStats/AVStats
    VotingStats.append([i+1,AVStats,AAVStats,percentage])


D = sorted(VotingStats, key=lambda student: student[3],reverse = True)

f = open("AdminVotingStats.txt",'w')
for i in D:
    f.write(str(i)+"\n")
f.close()    
```
